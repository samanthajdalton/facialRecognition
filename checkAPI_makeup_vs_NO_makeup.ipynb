```python

#Set toScroll = False for testing & fbPageName is name of club's fb page
def getPicUrls():

    url = 'http://sfglobe.com/?id=903'
    r = requests.get(url)
    tree = html.fromstring(r.text)
    
    listOfImageUrls = []    

    for htmlObject in tree.xpath("//img"): 
        urlOfImage = htmlObject.attrib['src']
        if urlOfImage.find('amazonaws') > 0:
            listOfImageUrls.append(urlOfImage)
            
    return listOfImageUrls

x = getPicUrls()
```
